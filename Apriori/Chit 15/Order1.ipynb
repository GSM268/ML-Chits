```python
df=pd.read_csv("Order1.csv")
```

```python
df
```

```python
groups=df.groupby(by="Member_number")
```

```python
records=groups.apply(lambda x:[x.iloc[i,2] for i in range(len(x))])
records
```

```python
association_rules=apriori(records,min_support=0.003,min_confidence=0.3,min_lift=3)
association_result=list(association_rules)
```

```python
print(association_result[0])
```

```python
for i in association_result:
  pair =i[0]
  item=[x for x in pair]

  print("Rules:"+item[0]+"---->"+item[1])
  print("Support"+str(i[1]))
  print("Confidence"+str(i[2][0][2]))
  print("Lift"+str(i[2][0][3]))
```
