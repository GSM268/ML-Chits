```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn import metrics
```

```python
#import dataset
```

```python
df = pd.read_csv("madfhantr.csv")
```

```python
df.head()
```

```python
#data preprocessing
```

```python
cols = df.columns
cols
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
var_mod = ['Loan_ID','Gender','Married','Dependents','Education','Self_Employed','Property_Area','Loan_Status']
```

```python
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i].astype(str))
```

```python
df.dtypes
```

```python
X = df[['Married', 'Dependents', 'Education','Self_Employed', 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount','Loan_Amount_Term', 'Credit_History', 'Property_Area']]
y = df[["Loan_Status"]]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=20)
```

```python
model = DecisionTreeClassifier().fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
plt.figure(figsize=(100,100))
tree.plot_tree(model,filled=True,fontsize=16)
plt.show()
```
